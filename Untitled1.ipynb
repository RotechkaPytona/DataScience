```python
import matplotlib.pyplot as plt

x1 = [91.4, 90, 89.7, 96.1, 87]
y1 = [38.5, 36.1, 37.3, 38.0, 28.4]
x2 = [91.3, 90.2, 89.8, 96.1, 86.9]
y2 = [29.6, 28.5, 29.0, 30.2, 24.8]


plt.title("Контроль")
plt.xlabel("Вес")
plt.ylabel("ИМТ")
plt.plot(x1, y1, marker='o', linestyle='', label='Данные на начало исследования')
plt.plot(x2, y2, marker='o', linestyle='', label='Данные на конец исследования')

plt.legend()

plt.show()
```
